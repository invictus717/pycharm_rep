```python
#matplotlib绘图基础；
import matplotlib.pyplot as plt
plt.figure(figsize=(3,2),facecolor="green")#创建一个背景是绿色，宽3英寸，高2英寸；
plt.plot()
plt.show()
```

```python
plt.rcParams["font.sans-serif"]="SimHei"
fig=plt.figure(facecolor="lightgrey")

plt.subplot(2,2,1)
plt.title("子标题1")
plt.subplot(2,2,2)
plt.title("子标题2",loc="left",color="b")
plt.subplot(2,2,3)
myfontdict={"fontsize":12,"color":"g","rotation":30}
plt.title("子标题3",fontdict=myfontdict)
plt.subplot(2,2,4)
plt.title("子标题4",color="white",backgroundcolor="black")
plt.suptitle("全局标题",fontsize=20,color="red",backgroundcolor="yellow")
plt.show()
```

```python
plt.rcParams["font.sans-serif"]="SimHei"
fig=plt.figure(facecolor="lightgrey")

plt.subplot(2,2,1)
plt.title("子标题1")
plt.subplot(2,2,2)
plt.title("子标题2",loc="left",color="b")
plt.subplot(2,2,3)
myfontdict={"fontsize":12,"color":"g","rotation":30}
plt.title("子标题3",fontdict=myfontdict)
plt.subplot(2,2,4)
plt.title("子标题4",color="white",backgroundcolor="black")
plt.suptitle("全局标题",fontsize=20,color="red",backgroundcolor="yellow")
plt.tight_layout()
plt.show()
```
